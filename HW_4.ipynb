Составим матрицу корреляций.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

_Вывод: самая сильная обратная корреляция между годом постройки и состоянием здания; сильная прямая корреляция между кадратурой и грейдом, ценой и количеством спален, ванн и жилой площадью._

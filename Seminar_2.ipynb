Чтобы вывести в нужной нам форме и два производителя, напишем лябда-функцию.

Функция лямбда задаеткакие именно индексы нужно взять - с максимальным значением - и передает их в Х - это уже число.

```python
df['Company'].value_counts().loc[lambda x: x==df['Company'].value_counts().max()]
```
